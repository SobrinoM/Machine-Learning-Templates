```python
#Remove missing data
#shift tab to inspect a function
from sklearn.impute import SimpleImputer

imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp = imp.fit(dataset.values[:,1:3])
# dataset.values[:,1:2]=imp.transform(dataset.values[:,1:3])
X.iloc[:,1:3] = imp.fit_transform(X.iloc[:,1:3])

#
display(X.head())
```

```python
#Encode catagorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
#labelencoder_X = LabelEncoder()
#X.iloc[:,0] = labelencoder_X.fit_transform(X.iloc[:, 0])

display(X.head())

#2>1 but spain>france --> OneHotEncoder
#from sklearn.compose import ColumnTransformer
#ct = ColumnTransformer([('encoder', OneHotEncoder(categories='auto'))], remainder='passthrough')
#X.iloc[0] = ct.fit_transform(X.iloc[0])

# Get dummies variable columns for categorical data; Drop first --> avoid dummy variable trap (multicoliniearity with the cte);
#To learn more on the dummy variable trap: https://www.algosome.com/articles/dummy-variable-trap-regression.html
#or on udemy: https://www.udemy.com/machinelearning/learn/v4/t/lecture/5772050?start=0
X = pd.get_dummies(X,columns=['Country'],drop_first=True)

display(X.head())

# Encoding Y data yes/No
from sklearn.preprocessing import LabelEncoder
Y = LabelEncoder().fit_transform(Y)
from IPython.display import display
display(Y)
```

```python
#Spliting the dataset to a test & training set
from sklearn.model_selection import train_test_split 

[X_train,X_test,Y_train,Y_test] = train_test_split(X, Y, test_size = 0.2, random_state = 1)
#X_train = X_train.astype(float)
display(X_train.head())
display(Y_train)
```

```python
# Feature Scaling
# Standardisation: normalisation & standardisation

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()

#Scaling 1-0 dummy variables: not needed, better for interpretation. Here we scale all:
X_train.iloc[:,0:2] = sc_X.fit_transform(X_train.iloc[:,0:2])
X_test.iloc[:,0:2] = sc_X.transform(X_test.iloc[:,0:2])

display(X_train.head())
display(X_test.head())
```
