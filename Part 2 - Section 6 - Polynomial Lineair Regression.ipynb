```python

# - - - - - - - - - - - -
# Polynomial Lineair Regression
# - - - - - - - - - - - -

#y = b_0 + b_1x_1 + b_2x_2^2 + b3_x_3^3
#Still linair, because the coëfficients are independent of each other (e.g. nonlinear: b2=b1x3/b2)


%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

# - Set path - 
#convert to raw string and add an extra \ to the end (not to escape the string)
os.chdir(r"C:\Users\msfernandez\Machine Learning A-Z\Machine Learning A-Z Template Folder\Part 2 - Regression\Section 6 - Polynomial Regression\\")

# - - - - - - - - - - - -
# - import the dataset - 
# - - - - - - - - - - - -
dataset = pd.read_csv("Position_Salaries.csv")
display(dataset.head())

X = dataset.iloc[: , 1:2] #Only Level column
Y = dataset.iloc[: , dataset.shape[1]-1]

# Compare it with a linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)

# - - - - - - - - - - - -
# Fitting Polynomial regression
# - - - - - - - - - - - -

from sklearn.preprocessing import PolynomialFeatures

#create a new feature set with x²: b + b1x1 + b2x2²
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
#Fit this poly feature set
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,Y)

#create a new feature set with x³: b + b1x1 + b2x2² + b3x3³
poly_reg3 = PolynomialFeatures(degree = 3)
X_poly3 = poly_reg.fit_transform(X)
#Fit this poly feature set
lin_reg3 = LinearRegression()
lin_reg3.fit(X_poly3,Y)

# - - - - - - - - - - - -
# Visualise
# - - - - - - - - - - - -

#Create a grid to plot the lines
X_grid = np.arange(min(X.values), max(X.values), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

#Create the plots
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
# Plot the linear regression results
ax1.scatter(X,Y, color ='red')
ax1.plot(X, lin_reg.predict(X))

# Plot the polynomial linear regression ² results
ax2.scatter(X,Y, color ='red')
ax2.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)))

# Plot the linear linear regression ³ results
ax3.scatter(X,Y, color ='red')
ax3.plot(X_grid, lin_reg3.predict(poly_reg.fit_transform(X_grid)))
```
