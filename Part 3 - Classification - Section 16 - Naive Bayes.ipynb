```python
# Naïve Bayes: 
# Given the features, what is the chance that the observation can be placed in a certain cluster.
# Look at sets of nearest neighbours to find the Marginal Likelihood. This is the ratio of "similar" events (vs total # events).
# denominator can be removes from bayes for a classifier (to compare values, not needed)

# Naïve because: independence between the features is implied for Bayes.


#libraries

%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

# - Set path - 
#convert to raw string and add an extra \ to the end (not to escape the string)
dir = (r'C:\Users\msfernandez\Machine Learning A-Z\Machine Learning A-Z Template Folder\Part 3 - Classification\Section 18 - Naive Bayes\\')
os.chdir(dir)
# - - - - - - - - - - - -
# - import the dataset - 
# - - - - - - - - - - - -
dataset = pd.read_csv('Social_Network_Ads.csv')
display(dataset.head())

X = dataset.iloc[:, [2,3]]
y = dataset.iloc[:, 4]

display(X.head())

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()

X_train = sc_X.fit_transform(X_train.astype(float))
X_test = sc_X.transform(X_test.astype(float))

#_______________________________________________________

# Fitting the classifier to the training set
from sklearn.naive_bayes import GaussianNB # (support vector classification)
classifier = GaussianNB()
classifier.fit(X_train, y_train)

#Predict the test set results
y_pred = classifier.predict(X_test)


# Making the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

display(cm)


```

```python
# Visualising the Test set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Naive Bayes (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
